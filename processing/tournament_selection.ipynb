```python
import pandas as pd

df = pd.read_csv("../data/tournament_ids.csv")

# extract date
df['name'] = df['name'].str.extract(r'(\d{1,2}\.\d{1,2}\.\d{4})')

# convert date
df['name'] = pd.to_datetime(df['name'], format='%d.%m.%Y')

# select tournaments between November and December
df = df[(df['name'].dt.month >= 11) & (df['name'].dt.month <= 12) & (df['name'].dt.year >= 2022)].reset_index(drop=True)

# display
df
```

```python
# select actually correct tournaments
df = df[df['tournament_id'].isin([57, 58, 59, 60, 61, 62, 102, 103, 104, 105, 106, 107, 152, 153, 154, 155, 156, 157])]
```

```python
# extract tournament IDs, these should contain all wanted tournament and some
tournament_ids = list(df['tournament_id'])
```

```python
import json

with open("../data/extracted_tournament_ids.json", "w") as f:
    json.dump(tournament_ids, f)
```
